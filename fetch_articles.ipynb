```python
import requests
import json
from tqdm import tqdm


url = "https://google.serper.dev/search"


data = []
articles = []
for i in tqdm(range(0, 100)):
  payload = {
    "q": "latest news articles on AI",
    "tbs": "qdr:w",
    "page": i
  }
  headers = {
    'X-API-KEY': '0c45362ce398eebae84a7c1937ec646ed33021dd',
    'Content-Type': 'application/json'
  }
  response = requests.request("POST", url, headers=headers, json=payload)

  page_articles = json.loads(response.text)

  if "organic" in page_articles:
    for art in page_articles["organic"]:
      art["position"] = i * 10 + art["position"]
    data.append(page_articles)
    articles.extend(page_articles["organic"])

len(articles)
```

```python
import json

with open("articles-150.jsonl", "w") as f:
    for article in articles:
        f.write(json.dumps(article) + "\n")
```
